## 研究活動のdomain

```javascript
var dom = {
    research : {},
    documentation : {},
    data: {},
    analysis : {},
    visualization : {},
    ideaProcessing : {},
    presentation : {},
    referenceManagement : {},
    versioning : {}
}

dom.research = new Domain({
    "title":"研究活動の構成要素とツール",
    "contents": [`= 研究活動の構成要素とツール\n${asciidocHeader}\n\ninclude::${srcdir}/dom_research.adoc[]`]
})

dom.documentation = new Domain({
    title:"Documentation",
    contents:[ `= Documentation\n${asciidocHeader}\n\ninclude::${srcdir}/dom_documentation.adoc[]`  ]
}).processedBy([
    tool.word
])

dom.data = new Domain({
    title:"Data Management",
    contents:[`= Data\n${asciidocHeader}\n\ninclude::${srcdir}/dom_data.adoc[]`]
}).using([
    data.dataset
]).processedBy([
    tool.excel,
    tool.jupyter
])

dom.analysis = new Domain({
    title:"Analysis",
    contents:[`= Analysis\n${asciidocHeader}\n\ninclude::${srcdir}/dom_analysis.adoc[]`   ]
}).using([
    data.dataset,
    data.sourcecode,
    data.markup
]).processedBy([
    tool.jupyter,
    tool.vscode,
    tool.rstudio
])

dom.visualization = new Domain({
    title:"Visualization",
    contents:[`= Visualization\n${asciidocHeader}\n\ninclude::${srcdir}/dom_visualization.adoc[]`   ]
}).using([
    data.sourcecode,
    data.dataset
]).processedBy([
    tool.jupyter,
    tool.vscode,
    tool.fumipos,
    tool.fgraph
])

dom.ideaProcessing = new Domain({
    title: "Idea processing",
    contents:[  `= Idea processing\n${asciidocHeader}\n\ninclude::${srcdir}/dom_ideaProcessing.adoc[]`   ]
}).using([
    data.markup
]).processedBy([
    tool.vscode,
    tool.jupyter,
    tool.fgraph
])

dom.presentation = new Domain({
    title:"Presentation",
    contents:[`= Presentation\n${asciidocHeader}\n\ninclude::${srcdir}/dom_presentation.adoc[]`]
}).processedBy([
    tool.fgraph,
    tool.powerpoint
])

dom.referenceManagement = new Domain({
    title:"Reference management",
    contents:[`= Reference manager\n${asciidocHeader}\n\ninclude::${srcdir}/dom_referenceManagement.adoc[]`]
})
dom.versioning = new Domain({
    title:"Versioning",
    contents:[`= Versioning\n${asciidocHeader}\n\ninclude::${srcdir}/dom_versioning.adoc[]`]
}).processedBy([
    tool.git
])





dom.research.composedOf([
    dom.documentation,
    dom.data,
    dom.analysis,
    dom.visualization,
    dom.ideaProcessing,
    dom.presentation,
    dom.versioning,
    dom.referenceManagement
])

```
