```javascript
var modelAssumption = {}

  modelAssumption.multiMixing = new MyPremise({
    title: "多段階のマグマ混合",
    contents: `
= 多段階のマグマ混合

[simple]
* 組成プロファイルの不連続はメルト組成が短い時間スケールで急に変化したことを示すと解釈可能

`
  }).consistentWith([data_for_crystal_growth.discontinuousProfile])



  modelAssumption.noCrustalAssimilation = new MyPremise({
    title: "地殻混染は無い",
    contents: `= 地殻混染は考慮しない

== 根拠

[simple]
- 解析対象のホスト溶岩はSiO~2~濃度が低いのでSiO~2~濃度の大きいメルトの影響は少ないと判断
- 地殻物質の外来結晶が著しく溶融した様子が見られない
  - 外来結晶として高頻度で見られる石英やカリ長石は地表近くのマサと予想
`
  })



  modelAssumption.multiCrystallization = new MyPremise({
    title: "多段階の結晶成長",
    contents: `= 多段階の結晶成長

== 仮定

- マグマ混合後のホストメルトから晶出開始
- ホストメルトは均質
- 再溶融・溶食なし
- 結晶成長のカイネティクスは考慮しない
`
  }).consistentWith([
    data_for_crystal_growth.correlationOfComponents,
    data_for_crystal_growth.euhedralZoning
  ])



  modelAssumption.multiDiffusion = new MyPremise({
    title: "multiple lattice diffusion",
    contents: `= 多段階の斑晶内元素拡散

== 仮定

- Cr~2~O~3~に比べFe/Mgプロファイルが均されているので元素拡散の影響がある
- 各成分の拡散速度は温度のみに依存

`
  }).consistentWith([data_for_crystal_growth.sharpnessOfProfiles])




  modelAssumption.crystallizingStoichiometry = new MyPremise({
    title: "Stoichiometry of crystallization",
    contents: `= 結晶化のストイキオメトリー

== 斑晶組み合わせ

[simple]
* Olivine
  ** > 2mm 単結晶 骸晶
  ** Opx集斑晶の中央部
* Orthopyroxene
  ** 自形 単結晶
  ** 半自形 olivineをともなう集斑晶
* Spinelはインクルージョンとして含まれる`
  }).consistentWith([data_for_crystal_growth.combinationOfPhenocrysts])



  modelAssumption.systemComponents = new MyPremise({
    title: "系の組成",
    contents: `= 系の組成

== SiO~2~ - MgO - FeO - Cr~2~O~3~ - Other system

[simple]
* Olivineとorthopyroxene
  ** 主要元素: SiO~2~・MgO・FeO
  ** 微量元素: Cr~2~O~3~
* Spinel
  ** Cr~2~O~3~のみ考慮
`
  }).consistentWith([data_for_crystal_growth.combinationOfPhenocrysts])




  modelAssumption.localEquilibrium = new MyPremise({
    title: "Local equilibrium",
    contents: `= 固液界面で局所平衡が成立したもとでの結晶成長

== 仮定

[simple]
* 結晶化の間結晶内部の組成は変化しない
* 元素拡散のときも局所平衡
`
  }).consistentWith([data_for_crystal_growth.euhedralZoning])



modelAssumption.timeScaleOfZoningFormation= new MyPremise({
    title: "Time scale of forming chemical zonig",
    contents:[
        `= 組成ゾーニング形成に関わるタイムスケール

[simple]
* 斑晶内拡散のタイムスケールは結晶成長タイムスケールより長い
* 結晶成長終了後に元素拡散が始まる
* マグマ混合の間は結晶成長も元素拡散も進行しない
        `
    ]
}).consistentWith([
    data_for_crystal_growth.timescale.growthRate,
    data_for_crystal_growth.timescale.diffusionRate,
    data_for_crystal_growth.timescale.mixingRate
])


// test
  var readme = new MyData({
    title: "Read me",
    contents: [
      "./readme.md",
      `${documentRoot}/toko/fitting_to_tanh.html`,
      `${documentRoot}/Articles/Basalt/Annen et al-2006-The genesis of intermediate and silicic magmas in deep crustal hot zones.pdf`
    ]
  })

  var readmeAdoc = new MyData({
    title: "AsciiDoc",
    contents: [`./readme.adoc`,
      `
# ここから2つめのコンテンツ(markdown literal)

## 複数のタイプのメディアをロードする

最初のコンテンツはasciidoc形式だったが, このコンテンツはmarkdown形式である
`,
      `./image/test_diagram.html`
    ]
  })

  var pdf = new MyData({
    title: "pdf",
    contents: `${documentRoot}/Articles/Basalt/Annen et al-2006-The genesis of intermediate and silicic magmas in deep crustal hot zones.pdf`
  })

  
  
  
  
// Model
  var model_crystallization = {};

model_crystallization.mixingModel = new MyModel({
    title:"Multiple magma mixing",
    contents:[
        `= Multiple magma mixing model`
    ]  
})
    .assuming([
     modelAssumption.multiMixing,
    modelAssumption.noCrustalAssimilation,
])




model_crystallization.growthModel = new MyModel({
    title:"Multiple growth",
    contents:[
        `= Multiple crystal growth model

== 概要

波動累帯構造の形成を表現するため, ホストメルト組成が不連続に変化することを仮定した

固相組成は液組成と分配係数のみで決まるとする
`        
    ]
})
    .assuming([
     modelAssumption.multiCrystallization,
     modelAssumption.crystallizingStoichiometry,
     modelAssumption.systemComponents,
     modelAssumption.localEquilibrium
])




model_crystallization.diffusionModel = new MyModel({
    title: "Multiple diffusion",
    contents:[
        `= Multiple elemental diffusion model

== 概要

* 結晶成長の後に元素拡散が開始
* 拡散進行の温度・タイムスケールを表す変数を導入

== 定式化

* 次元拡散方程式を空間方向, 時間方向に無次元化 (Lasaga, 1983)

$$
    \\begin{equation}
    \\frac{\\partial C_i}{\\partial t} = D(T(t)) \\frac{\\partial^2 C_i}{\\partial x^2}
    \\end{equation}
$$

* クランクｰニコルソン法により陰解法で数値計算
  ** 時間ステップは10分割

== 元素拡散の温度・時間タイムスケール

波動累帯構造が形成されるような状況においては, 冷却パスが複雑であるうえ, 温度の時間変化の推定が困難である.
本研究の目的は, 元素拡散の影響を受ける前の物質の分布が興味の対象であり, 具体的な冷却パスに関する仮定を増やしたくない.

結晶界面が時間とともに移動しないとき, 次のような変数を導入することで, 拡散係数の温度の陽な時間依存性を消すことができる.
ある時刻$t$までの拡散の進行度は, 拡散係数の時間積分によって得られる.
拡散係数が温度依存性を持つとき, 単位時間あたりの拡散の進行度は温度に依存する.
拡散進行後の量の空間分布を, 拡散の進行度で表すことを考える. 
拡散係数を初めの温度$T(t=0)$のときの拡散係数で規格化したものを, 積分したものを$\\tau(t)$と表すとすると, 

$$
    \\begin{equation}
    \\tau(t) = \\int^{t}_{0} \\left[ \\frac{D(T(\\hat{t}))}{D(0)} \\right]d\\hat{t}
    \\end{equation}
$$

ここで, 任意の異なる2つの時刻における温度を交換したとしても, $\\tau(t)$は変わらない.
また, このような操作を行っても時刻$t$までの拡散の進行度は変わらない. 
従って, $\\tau(t)$を用いることで拡散の進行度が等しくなるような, 全ての冷却パス$T(t)$を代表させることができる.

また, $\\tau(t)$は単調増加関数であり, その最大値が拡散終了時の量の空間分布に対応する.

本研究では, $\\tau(t)$の最大値$\\tau^{total}$を未知数として推定することにした.
これにより, 冷却パスが未知であっても元素拡散の影響を考慮することが可能になった. 

== 問題点

* 冷却パス$T(t)$が同じでも, 成分ごとに$\\tau^{total}$は異なるが, それが同一になっている
  ** 元素拡散の効果が過大または過小になっている成分がある
* 液相組成の時間変化を考慮していない
  ** 元素拡散の効果が過小になっている

`
    ]
})
.assuming([
    modelAssumption.multiDiffusion,
    modelAssumption.localEquilibrium
])



  model_crystallization.composedModel = new MyModel({
    title: "multiple mixing, crystal growth, and lattice diffusion model",
    contents: `${documentRoot}/thesis/decode_zoning/common_zoning_formation.adoc`
  })
    .composedOf([
      model_crystallization.mixingModel,
      model_crystallization.growthModel,
      model_crystallization.diffusionModel
  ])
    .assuming([
      modelAssumption.timeScaleOfZoningFormation
  ])
    .basedOn([
      pdf,
      readme,
      readmeAdoc
    ])

console.log(Object.keys(modelAssumption))
```
